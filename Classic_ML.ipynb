# ML Problems


```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans,DBSCAN
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
iris = load_iris(as_frame=True)
df = iris.frame
```

```python
df
```

```python
X=df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].copy()
Y=df['target']
```

## Supervised Problems


### KNN

### Logistic Regression

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
model= LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy2 = accuracy_score(y_test,y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,y_pred), display_labels = [0, 1,2])
cm_display.plot()
plt.show() 
```

### Random Forest

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
model= RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy3 = accuracy_score(y_test,y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,y_pred), display_labels = [0, 1,2])
cm_display.plot()
plt.show() 
```

### Descion Tree

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
model= DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy4 = accuracy_score(y_test,y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test,y_pred), display_labels = [0, 1,2])
cm_display.plot()
plt.show() 
```

## UnSupervised Problems

### Kmeans

```python
model = KMeans(n_clusters=3)
y_pred = model.fit_predict(X)
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=Y)
plt.show()
accuracy5 = accuracy_score(Y,y_pred)
## PLOT BASED ON ON Features 1,2
```

### DbScan Clustering

```python
model = DBSCAN(eps=0.5, min_samples=5)
y_pred = model.fit_predict(X)
accuracy6 = accuracy_score(Y,y_pred)
```

### PCA

```python
model = PCA(n_components=2)
X_d = model.fit_transform(X)
plt.scatter(X_d[:,0], X_d[:,1], c=Y)
plt.show()
accuracy7 = accuracy_score(Y,y_pred)
## PLOT BASED ON PC1 AND PC2 NOT ON Features
```
